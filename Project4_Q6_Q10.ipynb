Load Uber data into a DataFrame

```python
import networkx as nx
import pandas as pd
la = pd.read_csv(os.path.join(path, 'los_angeles-censustracts-2019-4-All-MonthlyAggregate.csv'))
```

Select for December

```python
la_dec = la[la['month'] == 12]
print(la_dec.head())
```

Merge duplicate edges by averaging their weights. Duplicate edges may exist when the dataset provides you with the statistic of a road in both directions. We remove duplicate edges for the sake of simplicity.

```python
import ast 
la_dec['ends'] = [str(sorted(list(a))) for a in zip(la_dec['sourceid'], la_dec['dstid'])]
la_dec = la_dec.groupby('ends').mean().reset_index()

ends_list = la_dec['ends'].apply(lambda x: ast.literal_eval(str(x)))
la_dec[['sourceid','dstid']] = pd.DataFrame(ends_list.tolist(), index= la_dec.index)
```

Convert DataFrame to graph

```python
g = nx.convert_matrix.from_pandas_edgelist(la_dec, 'sourceid', 'dstid', True)
# print(g.nodes)
```

Add centroids as a node attribute

```python
centroids = dict(zip(range(1, len(centroid_mat)+1), centroid_mat))
nx.set_node_attributes(g, centroids, "centroid")
```

```python
print(g.nodes[2715]['centroid'])
```

Keep the largest connected component of the graph

```python
Gcc = sorted(nx.connected_components(g), key=len, reverse=True)
G0 = g.subgraph(Gcc[0]).to_undirected()
```

# QUESTION 8: Determine what percentage of triangles in the graph (sets of 3 points on the map) satisfy the triangle inequality. You do not need to inspect all triangles, you can just estimate by random sampling of 1000 triangles.

```python
cycle_list = nx.cycle_basis(G0)
tri_list = [x for x in cycle_list if len(x)==3]
print(len(tri_list))
print(tri_list[0])
```

```python
# G0.edges[1030, 9]['mean_travel_time']
def is_triangle(graph, node_list):
    sides = [graph.edges[node_list[i%3], node_list[(i+1)%3]]['mean_travel_time'] for i in range(3)]
    smallest,medium,biggest = sorted(sides)
    return smallest+medium>=biggest and all(s>0 for s in sides)
    
```

```python
import random
# n = len(tri_list)
n = 1000
tri_sample = random.sample(tri_list, n)
summ = 0
for tri in tri_sample:
    summ += is_triangle(G0, tri)

print(summ * 100 / n, '% of triangles pass the triangle inequality')
```

# QUESTION 9: Find an upper bound on the empirical performance of the approximate algorithm: Ï = Approximate TSP Cost / Optimal TSP Cost

```python
multi = mst.to_directed()
```

```python
len(mst.edges)
len(multi.edges)
```

```python
eu_path = nx.eulerian_path(multi, 2472)
eu_list = list(eu_path)
```

```python
embedded_tour = []
for edge in eu_list:
    if edge[0] not in embedded_tour:
        embedded_tour.append(edge[0])
    if edge[1] not in embedded_tour:
        embedded_tour.append(edge[1])
print(embedded_tour)
```

```python
mst_cost = 0
for edge in mst.edges:
    mst_cost += mst.edges[edge]['mean_travel_time']
print(mst_cost)
```

```python
approx_cost = 0
edge = []
final_path = []
extra = 0
for i, node in enumerate(embedded_tour):
    edge.append(node)
    if i == 0:
        final_path.append(node)
        continue
    if not G0.has_edge(edge[0], edge[1]):
        new_path = nx.dijkstra_path(G0, edge[0], edge[1], weight='mean_travel_time')
        new_edge = []
        extra += len(new_path)-2
        for j, new_node in enumerate(new_path):
            new_edge.append(new_node)
            if j == 0:
                continue
            final_path.append(node)
            approx_cost += G0.edges[new_edge]['mean_travel_time']
            new_edge.pop(0)
        edge.pop(0)
        continue
    final_path.append(node)
    approx_cost += G0.edges[edge]['mean_travel_time']
    edge.pop(0)
print(approx_cost)
print(extra)
```
